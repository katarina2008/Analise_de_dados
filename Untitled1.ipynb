```python
# IMPORTAÇÃO DE BIBLIOTECAS
import pandas as pd #Importar e manipular bases de dados
import matplotlib.pyplot as plt #para plotar graficos na tela
import seaborn as sns #para plotar graficos na tela
```

```python
# CARREGAR OS DADOS
df = pd.read_csv("https://raw.githubusercontent.com/katarina2008/Analise_de_dados/refs/heads/main/house-prices-advanced-regression-techniques/train.csv")
df.head()
```

```python
corr_matrix = df.select_dtypes(include='number').corr()
plt.figure(figsize=(20, 25))  # Define o tamanho do gráfico
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriz de Correlação - Dados de Imóveis')
plt.show()
```

```python
# Filtra as variáveis com correlação acima de 0.5 com 'SalePrice'
saleprice_corr = corr_matrix['SalePrice'].sort_values(ascending=False)
significant_features = saleprice_corr[saleprice_corr.abs() > 0.5].index

```

```python
# Exibe as variáveis mais correlacionadas com 'SalePrice'
print("Variáveis mais correlacionadas com SalePrice:")
print(saleprice_corr[saleprice_corr.abs()>0.5])
```

```python
# Identifica pares de variáveis com alta correlação (acima de 0.8)
high_corr_pairs = [(var1, var2) for var1 in corr_matrix.columns for var2 in corr_matrix.columns
                   if var1 != var2 and abs(corr_matrix.loc[var1, var2]) > 0.8]

print("Pares de variáveis com alta correlação (> 0.8):")
for var1, var2 in high_corr_pairs:
    print(f"{var1} e {var2}")
```
